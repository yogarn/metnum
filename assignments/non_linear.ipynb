```python
from random import randint

def bisect(fx, a, b, e):
    print(f"a = {a}")
    print(f"b = {b}")
    print("a", "\t", "b", "\t", "f(a)", "\t", "f(b)", "\t", "c", "\t", "f(c)")
    c = (a + b)/2
    i = 0

    e = 0

    print("a", "\t", "b", "\t", "fx(a)", "\t", "fx(b)", "\t", "c", "\t", "fx(c)")
    if (fx(a) * fx(b) < 0):
        while(abs(fx(c)) > e):
            c = (a + b)/2
            i += 1
            print(round(a,2), "\t", round(b,2), "\t", round(fx(a),2), "\t", round(fx(b),2), "\t", round(c,2), "\t", round(fx(c),2))
            if(fx(c) == 0):
                break
            if(fx(a) * fx(c) < 0):
                b = c
            elif(fx(a) * fx(c) > 0):
                a = c
        print(f"nilai akar persamaan adalah {c}")
        print(f"i = {i}")
    else:
        print("a dan b tidak memenuhi syarat")
```

```python
bisect(lambda x: x**3 - 2*(x**2) + 6*x - 4, -1, 1, 1e-7)
```
